```python
#Step 4: Visualization
import matplotlib.pyplot as plt
```

```python
#Visualize the ratings column with a histogram!

# create figure and axis
figure, axis = plt.subplots()
# plot histogram
axis.hist(choko['Rating'])
# set title and labels
axis.set_title('Chocolate Dataset')
axis.set_xlabel('Ratings')
axis.set_ylabel('Frequency')
```

```python
#Convert the Column Percent
choko['Cocoa\nPercent'] = choko['Cocoa\nPercent'].str.rstrip('%').astype('float') / 100.0
```

```python
#Plot the converted numerical Cocoa Percent values against the Rating values

# create a figure with its axis
figure, axis = plt.subplots()
# scatter the “sepal length” on the x-axis and the “sepal width” on the y-axis
axis.scatter(choko['Cocoa\nPercent'], choko['Rating'])
# set a title and labels
axis.set_title('Chocolate Dataset')
axis.set_xlabel('percent')
axis.set_ylabel('ratings')
```

```python
# Normalization
choko['Rating']=(choko['Rating']-choko['Rating'].mean())/choko['Rating'].std()
print(choko['Rating'])
```

```python
#Step 7
#List the companies ordered by their average score (averaged over each company’s reviews)

choko.sort_values(['Company\nLocation','Rating' ]).set_index(['Company\nLocation','Rating'])
```
